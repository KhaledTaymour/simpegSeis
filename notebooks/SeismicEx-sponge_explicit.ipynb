```python
cs = 0.5
hx = np.ones(500)*cs
hy = np.ones(500)*cs
mesh = Mesh.TensorMesh([hx, hy], 'CC')
```

## Acoustic Wave equation in time domain (1st order form)

$ \rho\frac{\partial p}{\partial t} = \nabla \cdot \vec{u}+\frac{\partial s}{\partial t}\delta(\vec{r}-\vec{r}_s)$

$ \mu^{-1}\frac{\partial \vec{u}}{\partial t} = \nabla p$

## Add damping term

$ \rho[\frac{\partial p}{\partial t} + \sigma p] = \nabla \cdot \vec{u}+\frac{\partial s}{\partial t}\delta(\vec{r}-\vec{r}_s)$

$ \mu^{-1}[\frac{\partial \vec{u}}{\partial t}+\sigma \vec{u}] = \nabla p$

## Discretized form

$ \mathbf{diag}(\rho)\frac{\mathbf{p}^{n+1}-\mathbf{p}^{n}}{\triangle t} +  \mathbf{diag}(\rho\sigma)\mathbf{p}^n = \mathbf{Div} \mathbf{u}^{n+1/2}+\mathbf{M}^{cc -1}\frac{\mathbf{s}^{n+1}-\mathbf{s}^{n}}{\triangle t}$ 


$ \mathbf{diag}(\mathbf{Av}^{f}_{cc}\mu^{-1})\frac{\mathbf{u}^{n+1/2}-\mathbf{u}^{n-1/2}}{\triangle t} + \mathbf{diag}(\mathbf{Av}^{f}_{cc}\mu^{-1}\sigma)\mathbf{u}^{n+1/2}=  \mathbf{Grad} \mathbf{p}^{n}$

## Compute p first

$ \mathbf{p}^{n+1} = \mathbf{p}^{n} - \triangle t\mathbf{diag}(\sigma)\mathbf{p}^n + \mathbf{diag}(\rho)^{-1}\triangle t[\mathbf{Div} \mathbf{u}^{n+1/2}+\mathbf{M}^{cc -1}\frac{\mathbf{s}^{n+1}-\mathbf{s}^{n}}{\triangle t}]$ 



## Then compute u


$ \mathbf{u}^{n+1/2} = \mathbf{u}^{n-1/2} - \triangle t\mathbf{diag}(\sigma)\mathbf{u}^{n-1/2}+\triangle t\mathbf{diag}(\mathbf{Av}^{f}_{cc}\mu^{-1})^{-1}\mathbf{Grad} \mathbf{p}^{n}$

```python
mesh.setCellGradBC('dirichlet')
Grad = mesh.cellGrad
Div = mesh.faceDiv
rho = np.ones(mesh.nC)*2.7
vhalf = 2000
vblk = 2800
v = np.ones(mesh.nC)*vhalf
blkind = np.logical_and(mesh.gridCC[:,1]>-12.5, mesh.gridCC[:,1]<12.5) & np.logical_and(mesh.gridCC[:,0]>-12.5, mesh.gridCC[:,0]<12.5)
v[blkind] = vblk
mu = rho*v**2
AvF2CC = mesh.aveF2CC
```

```python
dat = mesh.plotImage(v)
plt.colorbar(dat[0])
```

```python
time = np.linspace(0, 0.08, 2**10)
dt = time[1]-time[0]
```

## Determine $\sigma$

$ \sigma = 2 \frac{1-f}{f} \triangle t^{-1} $

```python
ax = mesh.vectorCCx[-30]
ay = mesh.vectorCCy[-30]
indy = np.logical_or(mesh.gridCC[:,1]<=-ay, mesh.gridCC[:,1]>=ay)
indx = np.logical_or(mesh.gridCC[:,0]<=-ax, mesh.gridCC[:,0]>=ax)
tempx = zeros_like(mesh.gridCC[:,0])
tempx[indx] = (abs(mesh.gridCC[:,0][indx])-ax)**2
tempx[indx] = tempx[indx]-tempx[indx].min()
tempx[indx] = tempx[indx]/tempx[indx].max()
tempy = zeros_like(mesh.gridCC[:,1])
tempy[indy] = (abs(mesh.gridCC[:,1][indy])-ay)**2
tempy[indy] = tempy[indy]-tempy[indy].min()
tempy[indy] = tempy[indy]/tempy[indy].max()
temp = tempx+tempy
temp[temp>1.] = 1.
f = 1-temp*0.1
sig = (1.-f)/f*2./dt
```

```python
dat = mesh.plotImage(sig)
plt.colorbar(dat[0])
```

```python
plt.plot(mesh.vectorCCx, sig[mesh.gridCC[:,1]==mesh.vectorCCy[125]], '.')
```

```python
sigy = zeros_like(mesh.gridCC[:,1])
sigy[mesh.gridCC[:,1]<-ay] = (mesh.gridCC[:,1][mesh.gridCC[:,1]<-ay]-ay)**2
refy = sigy[mesh.gridCC[:,0]<-ay].max()
sigy[mesh.gridCC[:,1]>ay] = (mesh.gridCC[:,1][mesh.gridCC[:,1]>ay]+ay)**2
sigy = sigy/refy
```

```python
def ricker(fpeak, t, tlag):
    """
        Generating Ricker Wavelet
        
        .. math ::
        
        
    """
    return (1-2*np.pi**2*fpeak**2*(t-tlag)**2)*np.exp(-np.pi**2*fpeak**2*(t-tlag)**2)
```

```python
wave = ricker(400, time, 0.0025)
```

```python
fig, ax = plt.subplots(1,1, figsize = (7, 5))
ax.plot(time, wave, '.-')
ax.set_xlim(0, 0.02)
```

```python
txind = Utils.closestPoints(mesh, [0., 30.], gridLoc='CC')
q = Utils.sdiag(1/mesh.vol)*np.zeros(mesh.nC)
q[txind] = 1.
```

```python
p = np.zeros((mesh.nC, time.size))
pn = np.zeros(mesh.nC)
p0 = np.zeros_like(pn)
un = np.zeros(mesh.nF)
u0 = np.zeros_like(un)
```

```python
Proj = mesh.getInterpolationMat(np.r_[3., 30], 'CC')
```

```python
%%time
for i in range(time.size-1):
    sn = ricker(400, time[i+1], 0.0025)
    s0 = ricker(400, time[i], 0.0025)
    pn = p0-dt*Utils.sdiag(sig)*p0+Utils.sdiag(1/rho)*dt*(Div*un+(sn-s0)/dt*q)
    p0 = pn.copy()
    un = u0 - dt*Utils.sdiag(AvF2CC.T*sig)*u0 + dt*Utils.sdiag(1/(AvF2CC.T*(1/mu)))*Grad*p0
    u0 = un.copy()
    p[:,i+1] = pn
```

```python
data = Proj*p
```
